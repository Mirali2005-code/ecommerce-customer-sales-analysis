```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("ecommerce_data.csv")

# Convert timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date

# Funnel
funnel = df['event_type'].value_counts()

plt.figure()
funnel.plot(kind='bar')
plt.title("Funnel Analysis")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.show()

# Revenue over time
revenue = df[df['event_type']=='purchase'].groupby('date')['price'].sum()

plt.figure()
revenue.plot()
plt.title("Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()

# Top products
top_products = df[df['event_type']=='purchase'].groupby('product_id')['price'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_products.plot(kind='bar')
plt.title("Top Products")
plt.show()
```
